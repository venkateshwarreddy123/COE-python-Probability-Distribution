**1. Probability Mass Function (PMF)**

Used for Discrete Random Variables

The PMF gives the probability of a discrete random variable X taking a specific value x:

P(X=x)=f(x)

Rolling a Fair 6-Sided Die

The possible values of X are 1, 2, 3, 4, 5, 6.

Since each outcome is equally likely:

P(X=x)= 1/6 ,for x∈{1,2,3,4,5,6}

**Python Code for PMF**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import binom

# Define values of X (die outcomes)
x_values = np.arange(1, 7)
pmf_values = [1/6] * 6  # Each value has equal probability

# Plot the PMF
plt.figure(figsize=(8, 4))
plt.bar(x_values, pmf_values, color='b', alpha=0.6, label="PMF")
plt.xlabel("Outcome (X)")
plt.ylabel("Probability")
plt.title("PMF of a Fair 6-Sided Die")
plt.xticks(x_values)
plt.legend()
plt.show()
```

**Bernoulli Distribution**

The Bernoulli distribution is a discrete probability distribution for a random variable that takes only two possible outcomes:

   **1.Success (1) with probability p**

   **2.Failure (0) with probability (1 - p)**

This distribution is useful in scenarios where an event either occurs (1) or does not occur (0).

**Mathematical Definition**

If X is a Bernoulli-distributed random variable with probability p, then:

**P(X = 1) = p , P(X = 0) = 1-p**

The probability mass function (PMF) is:

P(X = x) = p^x (1-p)^1-x , x {0,1}

**where:**

p = probability of success

1−p = probability of failure

**Properties:**

Mean (Expected Value): 
E[X]=p

Variance: 
Var(X)=p(1−p)

***Applications of Bernoulli Distribution:***

The Bernoulli distribution is widely used in real-world applications, such as:

**1.Medical Trials: Whether a patient responds to a treatment (1) or not (0).**

**2.Spam Detection: Whether an email is spam (1) or not spam (0).**

**3.Marketing: Whether a user clicks on an ad (1) or ignores it (0).**

**4.Quality Control: Whether a manufactured product is defective (1) or non-defective (0).**

**Example Dataset**

Let’s consider a medical trial where we test whether a new drug works or not. We conduct 10 trials, and for each trial:

1 → Patient recovers (Success)

0 → Patient does not recover (Failure)

p = Number of Successes/Total Triass

p=0.6 (Success probability)

Implementing Bernoulli Distribution in Python:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli

# Define probability of success
p = 0.6  

# Generate 1000 Bernoulli samples
data = bernoulli.rvs(p, size=1000)

# Plot the distribution
plt.bar([0, 1], [sum(data == 0), sum(data == 1)], color=['red', 'green'], tick_label=["Failure (0)", "Success (1)"])
plt.xlabel("Outcome")
plt.ylabel("Frequency")
plt.title("Bernoulli Distribution (p=0.6)")
plt.show()
```

4.Can be simulated and visualized using Python.

**Implementing Binomial Distribution in Python**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Define parameters
n = 10   # Total trials
p = 0.3  # Probability of success

# Generate Binomial distribution data
x = np.arange(0, n+1)
pmf = binom.pmf(x, n, p)

# Plot the Binomial Distribution
plt.bar(x, pmf, color='blue', alpha=0.7)
plt.xlabel("Number of Successes (Clicks)")
plt.ylabel("Probability")
plt.title(f"Binomial Distribution (n={n}, p={p})")
plt.show()
```

**Implementing Normal Distribution in Python**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define parameters
mu = 70   # Mean
sigma = 10  # Standard deviation

# Generate Normal distribution data
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)  # Range of scores
pdf = norm.pdf(x, mu, sigma)  # Compute PDF

# Plot the Normal Distribution
plt.plot(x, pdf, color='red', label=f"N({mu}, {sigma}²)")
plt.xlabel("Exam Scores")
plt.ylabel("Probability Density")
plt.title(f"Normal Distribution (μ={mu}, σ={sigma})")
plt.legend()
plt.grid()
plt.show()
```

4.The 68-95-99.7 Rule explains how data is distributed around the mean.

**Implementing Log-Normal Distribution in Python**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import lognorm

# Define parameters for Log-Normal Distribution
shape = 0.5  # σ (standard deviation of log(X))
scale = np.exp(10)  # e^μ (mean of log(X))

# Generate data points
x = np.linspace(0, 200000, 1000)
pdf = lognorm.pdf(x, shape, scale=scale)

# Plot Log-Normal Distribution
plt.figure(figsize=(8,5))
plt.plot(x, pdf, color='blue', label="Log-Normal Distribution")
plt.xlabel("Salary ($)")
plt.ylabel("Probability Density")
plt.title("Log-Normal Distribution of Salaries")
plt.axvline(scale, color='red', linestyle='--', label="Median Salary")
plt.legend()
plt.grid()
plt.show()
```

**Key Takeaways**

4.It is used in finance, economics, reliability analysis, and ML.

5.Transformation tip: If data is log-normal, applying log transformation can normalize it for analysis.

**Central Limit Theorem**

The Central Limit Theorem states that:

"If we take many random samples (each of size n) from any population with any distribution, the sampling distribution of the sample mean will be approximately normal if n is sufficiently large."

This means that even if the original data is skewed or non-normal, the means of samples will tend to be normally distributed as n increases.

1.Confidence Intervals & Hypothesis Testing – Used in A/B testing, medical trials, etc.

2.Stock Market Returns – Daily returns of assets follow normality due to CLT.

3.Manufacturing & Quality Control – Helps in setting tolerances in production.

4.Machine Learning – Feature distributions, data transformations, and modeling assumptions rely on CLT.

5.Polling & Surveys – Sample mean approximates the population mean when sample size is large.

**Demonstrating CLT in Python**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set random seed for reproducibility
np.random.seed(42)

# Generate a skewed income dataset (exponential distribution)
income_data = np.random.exponential(scale=50, size=1000)

# Plot the original skewed distribution
plt.figure(figsize=(8, 4))
sns.histplot(income_data, bins=50, kde=True, color="r")
plt.title("Original Income Distribution (Skewed)")
plt.xlabel("Income ($1000)")
plt.ylabel("Frequency")
plt.show()

# Perform sampling and calculate means
sample_means = []
n_samples = 1000  # Number of samples
sample_size = 30  # Sample size

for _ in range(n_samples):
    sample = np.random.choice(income_data, size=sample_size)
    sample_means.append(np.mean(sample))

# Plot the sampling distribution of the sample mean
plt.figure(figsize=(8, 4))
sns.histplot(sample_means, bins=50, kde=True, color="b")
plt.title(f"Sampling Distribution of Sample Mean (n={sample_size})")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```

**Key Observations**

1.Original Data is Skewed – Income data follows an exponential distribution.

2.Sample Means Follow Normal Distribution – Even though the original dataset is not normal.

3.Larger Samples Improve Approximation – As n increases, the distribution of sample means becomes more normal.
